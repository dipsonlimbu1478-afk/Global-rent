```python

import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
# Values
values = [6,2,6,6,3,6,1,6,4,3,2,6,1,6,5,4,6,6,3,4]
# Create histogram. Difine number of bins and colours.
plt.hist(values, bins=12, color='skyblue', edgecolor='black')
# Add labels and title
plt.title('Histogram of values')
plt.xlabel('values')
plt.ylabel('Frequency')
# Show plot
plt.show()
2

mean = np.mean(values)
variance = np.var(values)
std_dev = np.std(values)
mode=np.bincount(values).argmax()
print("mode:",mode)
print("mean:", mean)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```
